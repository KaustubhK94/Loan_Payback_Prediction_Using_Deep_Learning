### the Total_acc feature correlates with the mort_acc..

### Trying Fillna apporach...
### Group the dataframe with total accounts and calculate mean value for the mortage accounts per total_account entry

### atleast 10% of our values are missing  on this mort_acc column....
### so we cannot drop the rows else we'll drop 10 perc of our data.... 

```python
df.groupby("total_acc").mean()
```

```python
total_acc_mean =df.groupby("total_acc").mean()["mort_acc"]
```

```python
total_acc_mean
```

### so now we can use this series (above)....and fill in the missing values in the mort_acc 
### based on the average mort_acc values per total_acc...

```python
def fill_mort_acc(total_acc,mort_acc):
    if np.isnan(mort_acc):
        return total_acc_mean[total_acc]
    else:
        return mort_acc  
```

```python
df["mort_acc"] = df.apply(lambda x:fill_mort_acc(x["total_acc"],x["mort_acc"]),axis=1)
```

```python
df.isnull().sum()
```

### Seperation between dependant and independant features

```python
X = df.drop("loan_repaid",axis=1).values
```

```python
X
```

```python
y = df["loan_repaid"].values
```

```python
y
```

### since The target variables are imbalanced....perfoming Undersampling on dataset

```python
from imblearn.under_sampling import NearMiss
```

```python
nm = NearMiss()
X_res,Y_res = nm.fit_resample(X,y)
```

```python
X_res.shape,Y_res.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res,Y_res, test_size=0.20, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
X_train = scale.fit_transform(X_train)
```

```python
X_test = scale.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
## No. of neurons in first layer should match your no. of features
model.add(Dense(78,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(39,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(19,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(units = 1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
early_stop = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=20)
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=256,epochs=100,callbacks= [early_stop])
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss
```

```python
loss.plot()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions = model.predict_classes(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
# from sklearn.metrics import plot_confusion_matrix
# plot_confusion_matrix(model, X_test, y_test)  
# plt.show()
```

```python
print(confusion_matrix(y_test,predictions))
```
